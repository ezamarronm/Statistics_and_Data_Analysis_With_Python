```python
import pandas as pd
import numpy as no
import seaborn as sns
from scipy.stats import norm
```

```python
sample = [42, 35, 29, 45, 41, 57, 54, 47, 48, 56, 47, 35, 52, 31, 52, 55, 57, 58, 26, 29, 32, 37, 32, 34, 48, 20, 48, 51, 27, 24, 39, 40, 31, 34, 23, 24, 41, 58, 44, 48, 31, 23, 27, 55, 43, 47, 30, 57, 38, 51]
sample = pd.DataFrame(sample)
```

```python
mu1 = 37
mu2 = 42

std_dev = 2
mean = sample.mean()[0]

z_1 = (mean - mu1) / std_dev 
z_2 = (mean - mu2) / std_dev
```

```python
N = 1000000
data1 = norm.rvs(loc = mu1, scale = std_dev, size = N)
data2 = norm.rvs(loc = mu2, scale = std_dev, size = N)
```

```python
ax = sns.distplot(data1, bins = 500, kde = True, color = 'blue')
ax = sns.distplot(data2, bins = 500, kde = True, color = 'red')
ax.axvline(x = mean, label = 'Saample mean', color = 'black', linestyle = '--')
ax.legend()
```

```python
# Error 1 P(reject h0 | h0 is true)
mean_p = 1 - norm.cdf(z_1)
mean_p
```

```python
# Error 1 P(accept h0 | h0 is false)
mean_p = norm.cdf(z_2)
mean_p
```
