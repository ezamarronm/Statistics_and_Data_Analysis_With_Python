```python
data1 = t.rvs(100, size = 1000000) #Degrees of freedom, size
data2 = t.rvs(4, size = 1000000)
ax = sns.distplot(data1,bins = 500, kde = False, color= 'red')
ax = sns.distplot(data2,bins = 500, kde = False, color= 'blue')
```

```python
data1 = chi2.rvs(5, size = 1000000) #Degrees of freedom, size
data2 = chi2.rvs(15, size = 1000000)
ax = sns.distplot(data1,bins = 500, kde = False, color= 'red')
ax = sns.distplot(data2,bins = 500, kde = False, color= 'blue')
```

# Central limit theorem

```python
population = pd.DataFrame()
population['numbers'] = expon.rvs(40,size = 100000) #lambda, size
```

```python
population['numbers'].hist(bins = 100)
```

```python
size = 50000
n = 50
sample_mean = [population.sample(n = n).mean().values[0] for i in range(size)] #Getting n samples and appending its mean
```

```python
fig, ax = plt.subplots()
ax.hist(sample_mean, bins = 50, alpha = 0.5)
ax.set_xlabel('Mean Value')
ax.set_ylabel('Frequency')
ax.grid()
```
