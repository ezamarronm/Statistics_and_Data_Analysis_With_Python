```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats 

%matplotlib inline
```

```python
df = pd.read_csv('./shared_bicycles.csv')
```

```python
df.columns
```

```python
# Categoricals
y = df['cuartil-ano'].apply(lambda x: 'Quartile-' +str(int(x)))
fig, axes  = plt.subplots()
axes.bar(y.value_counts().index, y.value_counts())
axes.set_xlabel('Yearly Quartiles')
axes.set_ylabel('Frequency')
```

# Categoricals


```python
y = df['cuartil-ano'].apply(lambda x: 'Quartile-' +str(int(x)))
fig, axes  = plt.subplots()
axes.bar(y.value_counts().index, y.value_counts())
axes.set_xlabel('Yearly Quartiles')
axes.set_ylabel('Frequency')
axes.patches[2].set_facecolor('red')
axes.set_title('Bar Diagram')
```

```python
fig, axes = plt.subplots()
axes.pie(y.value_counts(), labels = y.value_counts().index)
axes.set_title('Pie Diagram')
```

# Numericals

```python
y = df['viento']
```

```python
fig,axes = plt.subplots()
axes.hist(y, bins = 30)
axes.set_xlabel('Wind')
axes.set_ylabel('Frequency')
plt.axvline(np.mean(y), c = 'red', linestyle = '--', label = 'Mean' )
plt.axvline(np.mean(y) + np.std(y), c = 'black', linestyle = '--', label = '+1 Standard Deviation')
plt.axvline(np.mean(y) - np.std(y), c = 'black', linestyle = '--', label = '-1 Standard Deviation' )
axes.legend()
```
